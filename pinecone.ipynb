```python
import os
from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv(), override=True)
api_key=os.getenv("PINECONE_API_KEY")
print(api_key)
from pinecone import Pinecone

# Initialize Pinecone client
pinecone_client = Pinecone(api_key=api_key)
# Check connection by listing indexes
try:
    indexes = pinecone_client.list_indexes()
    print("Connected successfully. Current indexes:", indexes)
except Exception as e:
    print("Failed to connect to Pinecone:", str(e))
```

```python
pinecone_client.list_indexes()
```

```python
from pinecone import PodSpec
from pinecone import Pinecone, ServerlessSpec
index_name = 'langchain'
if index_name not in pinecone_client.list_indexes():
    print(f"Index '{index_name}' not found. Creating index...")
    pinecone_client.create_index(name=index_name, metric='cosine',dimension=1536, spec=ServerlessSpec(
        cloud="aws",
        region="us-east-1"
    ) )

    print(f"Index '{index_name}' created.")

else:
    print(f"Index '{index_name}' already exists.")
```

```python
index = pinecone_client.Index(index_name)
index.describe_index_stats()
```

```python
import random
vectors = [[random.random() for _ in range(1536)] for v in range(5)]
print(vectors)
ids = list('abcde')
index_name = 'langchain'
index = pinecone_client.Index(index_name)
index.upsert(vectors=zip(ids, vectors))
```

```python
#updating the vectors
index.upsert(vectors=[('c', [0.5] * 1536)]) 
```

```python
#fetching the vectors
index.fetch(ids=['c','d'])
```

```python
#deleting the vectors
index.delete(ids=['c','b','d'])
```

```python
index.describe_index_stats()
```
